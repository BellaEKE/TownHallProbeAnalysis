###MERGE TABLES


```python
import pandas as pd

# Read in the tables
bunnies = pd.read_csv('Bunnies2.csv')
people = pd.read_csv('People.csv')
meetings = pd.read_csv('Meetings.csv')

# Merge the bunnies and meetings tables
bunnies_meetings = pd.merge(bunnies, meetings, left_on='id', right_on='bunny_id')

# Merge the people and meetings tables
people_meetings = pd.merge(people, meetings, left_on='id', right_on='person_id')

# Merge the bunnies/meetings and people/meetings tables
final_table = pd.merge(bunnies_meetings, people_meetings, on='meeting_id')

```

```python
df=final_table
df
```

# look for patterns or commonalities among the bunnies who have committed crimes by analyzing their breed or age.

We can start by grouping the crime_data by bunny breed and calculating the total number of crimes committed by each breed. This can be done using the groupby and sum functions:

```python
crimes_by_breed = crime_data.groupby('breed')['commited_crime'].sum()
crimes_by_breed
```

```python
crimes_by_breed2 = crime_data.groupby('breed')['commited_crime'].value_counts().sort_values(ascending=False)
crimes_by_breed2
```

# Similarly, we can group the crime_data by bunny age and calculate the total number of crimes committed by each age:

```python
crimes_by_age = crime_data.groupby('bunny_age')['commited_crime'].value_counts().sort_values(ascending=False)
crimes_by_age
```

# is there a relationship between the breed and crime count:do certain breeds have more tendency to commit crimes

```python
import matplotlib.pyplot as plt

# Create a bar plot of crime count by breed
plt.figure(figsize=(10,6))
crime_data.groupby('breed')['commited_crime'].sum().sort_values().plot(kind='barh')
plt.xlabel('Number of Crimes')
plt.ylabel('Breed')
plt.title('Crimes Committed by Breed')
plt.show()
```

# # is there a relationship between the age and crime count:do certain age have more tendency to commit crimes

```python
# Create a bar plot of crime count by age
plt.figure(figsize=(10,6))
crime_data.groupby('bunny_age')['commited_crime'].sum().sort_values().plot(kind='barh')
plt.xlabel('Number of Crimes')
plt.ylabel('bunny_age')
plt.title('Crimes Committed by Age')
plt.show()
```
